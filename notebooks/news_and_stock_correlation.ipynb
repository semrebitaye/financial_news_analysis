#### Date Alignment

Load datasets of news and each stock

```python
import pandas as pd

news_df = pd.read_csv('../Data/raw_analyst_ratings.csv')
apple_stock_df = pd.read_csv('../Data/yfinance_data/AAPL_historical_data.csv')
amazon_stock_df = pd.read_csv('../Data/yfinance_data/AMZN_historical_data.csv')
google_stock_df = pd.read_csv('../Data/yfinance_data/GOOG_historical_data.csv')
meta_stock_df = pd.read_csv('../Data/yfinance_data/META_historical_data.csv')
msft_stock_df = pd.read_csv('../Data/yfinance_data/MSFT_historical_data.csv')
nvda_stock_df = pd.read_csv('../Data/yfinance_data/NVDA_historical_data.csv')
tsla_stock_df = pd.read_csv('../Data/yfinance_data/TSLA_historical_data.csv')
```

Rename the Date Column of each stock to 'date'

```python
apple_stock_df.rename(columns={'Date': 'date'}, inplace=True)
amazon_stock_df.rename(columns={'Date': 'date'}, inplace=True)
google_stock_df.rename(columns={'Date': 'date'}, inplace=True)
meta_stock_df.rename(columns={'Date': 'date'}, inplace=True)
msft_stock_df.rename(columns={'Date': 'date'}, inplace=True)
nvda_stock_df.rename(columns={'Date': 'date'}, inplace=True)
tsla_stock_df.rename(columns={'Date': 'date'}, inplace=True)
```

Normalize dates

```python
# Convert to datetime, handling mixed formats and timezones
news_df['date'] = pd.to_datetime(news_df['date'], errors='coerce')

# Convert to UTC and then extract only the date part
news_df['date'] = news_df['date'].dt.tz_convert('UTC').dt.date
apple_stock_df['date'] = pd.to_datetime(apple_stock_df['date']).dt.date
amazon_stock_df['date'] = pd.to_datetime(amazon_stock_df['date']).dt.date
google_stock_df['date'] = pd.to_datetime(google_stock_df['date']).dt.date
meta_stock_df['date'] = pd.to_datetime(meta_stock_df['date']).dt.date
msft_stock_df['date'] = pd.to_datetime(msft_stock_df['date']).dt.date
nvda_stock_df['date'] = pd.to_datetime(nvda_stock_df['date']).dt.date
tsla_stock_df['date'] = pd.to_datetime(tsla_stock_df['date']).dt.date
```

Merge the datasets on the date column for each stock

```python
apple_merged_df = pd.merge(news_df, apple_stock_df, on='date', how='inner')
amazon_stock_df = pd.merge(news_df, amazon_stock_df, on='date', how='inner')
google_stock_df = pd.merge(news_df, google_stock_df, on='date', how='inner')
meta_stock_df = pd.merge(news_df, meta_stock_df, on='date', how='inner')
msft_stock_df = pd.merge(news_df, msft_stock_df, on='date', how='inner')
nvda_stock_df = pd.merge(news_df, nvda_stock_df, on='date', how='inner')
tsla_stock_df = pd.merge(news_df, tsla_stock_df, on='date', how='inner')
```
