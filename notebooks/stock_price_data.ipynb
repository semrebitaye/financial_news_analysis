### Apply Analysis Indicators with TA-Lib

##### 1, Amazon

```python
# Moving Averages
df_amazon['SMA_20'] = ta.SMA(df_amazon['Close'], timeperiod=20)  # 20-day Simple Moving Average
df_amazon['EMA_20'] = ta.EMA(df_amazon['Close'], timeperiod=20)  # 20-day Exponential Moving Average

# Relative Strength Index (RSI)
df_amazon['RSI_14'] = ta.RSI(df_amazon['Close'], timeperiod=14)  # 14-day RSI

# Moving Average Convergence Divergence (MACD)
df_amazon['MACD'], df_amazon['MACD_Signal'], df_amazon['MACD_Hist'] = ta.MACD(df_amazon['Close'], 
                                                               fastperiod=12, 
                                                               slowperiod=26, 
                                                               signalperiod=9)

# Display the first few rows to ensure the indicators are calculated correctly
print(df_amazon.head())
```

### Use PyNance for Financial Metrics

##### 1, Amazon

```python
# Calculate daily returns
df_amazon['Returns'] = df_amazon['Close'].pct_change()

# Calculate rolling volatility (e.g., 20-day rolling volatility)
df_amazon['Volatility_20'] = df_amazon['Returns'].rolling(window=20).std() * np.sqrt(252)

# Display the first few rows with the new metrics
print(df_amazon.head())
```

### Visualize the Data

##### 1, Amazon

```python
# Plotting the stock price along with moving averages
plt.figure(figsize=(14, 7))
plt.plot(df_amazon['Close'], label='Close Price')
plt.plot(df_amazon['SMA_20'], label='SMA 20')
plt.plot(df_amazon['EMA_20'], label='EMA 20')
plt.title('Stock Price with Moving Averages')
plt.legend()
plt.show()

# Plotting RSI
plt.figure(figsize=(14, 7))
plt.plot(df_amazon['RSI_14'], label='RSI 14')
plt.axhline(70, color='red', linestyle='--')
plt.axhline(30, color='green', linestyle='--')
plt.title('Relative Strength Index (RSI)')
plt.legend()
plt.show()

# Plotting MACD
plt.figure(figsize=(14, 7))
plt.plot(df_amazon['MACD'], label='MACD')
plt.plot(df_amazon['MACD_Signal'], label='MACD Signal')
plt.bar(df_amazon.index, df_amazon['MACD_Hist'], label='MACD Histogram')
plt.title('MACD')
plt.legend()
plt.show()

# Plotting Volatility
plt.figure(figsize=(14, 7))
plt.plot(df_amazon['Volatility_20'], label='20-Day Rolling Volatility')
plt.title('Stock Price Volatility')
plt.legend()
plt.show()
```
